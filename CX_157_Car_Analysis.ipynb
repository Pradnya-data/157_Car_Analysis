Handle Missing Values

```python
# Check for missing values

df.isnull().sum()
```

```python
df['AirBags']=df['AirBags'].fillna('Driver only')
```

```python
df['AirBags'].isnull().sum()
```

```python
# replace missing numerical value with mean
df['Rear.seat.room']=df['Rear.seat.room'].fillna(df['Rear.seat.room'].mean())
```

```python
df['Rear.seat.room'].isnull().sum()
```

```python
df['Luggage.room']=df['Luggage.room'].fillna(df['Luggage.room'].mean())
```

```python
df['Manufacturer'].mode()[0]
```

```python
#

for i in df.columns:
  if df[i].dtype=='object':
    m=df[i].mode()[0]
    df[i]=df[i].fillna(m)
  else:
    me=df[i].mean()
    df[i]=df[i].fillna(me)
```
